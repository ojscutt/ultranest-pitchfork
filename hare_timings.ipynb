```python
# stock imports
import numpy as np
import pandas as pd
import json
import scipy
import os
import random

##plotting
import matplotlib.pyplot as plt

# plt.style.use('dark_background')
plt.style.use("Solarize_Light2")
plt.rcParams.update({"axes.edgecolor": "black"})
plt.rcParams.update({"text.color": "black"})
plt.rcParams.update({"axes.labelcolor": "black"})
plt.rcParams.update({"xtick.color": "black"})
plt.rcParams.update({"ytick.color": "black"})
plt.rcParams.update({"font.family": "monospace"})
```

```python
m_times_dict = {}
for obs_idx in range(5):
    m_times=np.array([])
    for hare_idx in range(100):
        path = f'form/hare{hare_idx}/obs{obs_idx}/samples.pkl'
        m_times = np.append(m_times,os.path.getmtime(path))

    m_times_dict[f'm_times{obs_idx}'] = m_times
```

```python
plt.figure(figsize=(20,5))
for obs_idx in range(5):
    m_times = m_times_dict[f'm_times{obs_idx}']
    plt.plot(np.arange(1,100),m_times[1:]-m_times[:-1])

labels = ['initial_mass', 'initial_Zinit', 'initial_Yinit', 'initial_MLT', 'star_age', 'a', 'b']

plt.figure(figsize=(20,5))
for label in labels:
    
    hare_vals = np.array([])
    for hare_idx in range(100):
        hare_json = pd.read_json(f'form/hare{hare_idx}/hare{hare_idx}.json')
        val = hare_json[label]
        hare_vals =np.append(hare_vals, val)
    
    plt.plot(np.arange(1,100),hare_vals[1:])
```
