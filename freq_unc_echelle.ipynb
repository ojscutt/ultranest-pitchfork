```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
min_nu = 2800
delta_nu = 136

nus = [min_nu + delta_nu * i for i in range(15)]

plt.scatter([nu%delta_nu for nu in nus], nus)
```

```python
def obs_noise(true, unc, seed=None):
    seeded_random_state = np.random.RandomState(seed=seed)
    rvs_random_states = seeded_random_state.randint(0, high=2**32 -1,size=len(true))
    noisy_obs = np.empty(len(true))
    idx = 0
    for ob in true:
        noisy_obs[idx] = scipy.stats.norm(loc=ob, scale=unc[idx]).rvs(random_state=rvs_random_states[idx])
        idx+=1

    return noisy_obs

frequency_unc = 1

obs_unc = [frequency_unc+abs(i-25)*0.1 for i in range(18,33)]

nus_unc = obs_noise(nus, obs_unc)
nus_unc
```

```python
plt.scatter([nu%delta_nu for nu in nus], nus)
plt.scatter([nu%delta_nu for nu in nus_unc], nus_unc)
plt.xlim(0,delta_nu)
```
