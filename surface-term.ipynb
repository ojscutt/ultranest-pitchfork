```python
#stock imports
import numpy as np
import pandas as pd
import json
import scipy
import os

import matplotlib.pyplot as plt
```

```python
## dumb sim of mode freqs
n = np.arange(15,25)
dnu = 135
eps = .35
nu_max = 3050
freqs = (n+eps)*dnu
```

```python
def surf_corr(freqs, nu_max, a, b):
    return freqs + a*((freqs/nu_max)**b)
```

```python
plt.scatter(freqs%dnu, freqs)

corr_freqs = surf_corr(freqs,nu_max, -4.73, 4.9)
plt.scatter(corr_freqs%dnu, corr_freqs)
plt.xlim((0,dnu))
```

```python
# varying a and b

n = np.arange(15,25)
dnu = 135
eps = .35
nu_max = 3050
freqs = (n+eps)*dnu

n_val = 30
a_vals = np.linspace(-10, 2, n_val)
b_vals = np.linspace(4.4, 5.25, n_val)

a, b = np.meshgrid(a_vals, b_vals)

for i in range(n_val):
    for j in range(n_val):
        a_val = a[i,j]
        b_val = b[i,j]
        corr_freqs = surf_corr(freqs,nu_max, a_val, b_val)
        plt.plot(corr_freqs%dnu, corr_freqs, label=f'a={a_val}, b={b_val}', c='orange', alpha=0.01)

plt.scatter(freqs%dnu, freqs)
plt.xlim((0,dnu))
plt.show()
```
